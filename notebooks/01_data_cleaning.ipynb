```python
####################################################
# 1.13.2 Proxy - Anteile Wohnungstypen nach Gemeinde (oppervlakte_totaal_egw_mgw_gemeente_2019_2023.csv)
####################################################

df_woning_type_raw = pd.read_csv("../data_raw/oppervlakte_totaal_egw_mgw_gemeente_2019_2023.csv", sep=";")
print(df_woning_type_raw.shape)
df_woning_type_raw.head(10)
```

```python
# Kopie für sauberes Arbeiten
df_wt = df_woning_type_raw.copy()

# Jahr extrahieren
df_wt["year"] = df_wt["Perioden"].str[:4].astype(int)

# EGW extrahieren
df_egw = df_wt[df_wt["Woningtype"] == "ZW10290"].rename(
    columns={
        "BeginstandWoningvoorraad_1": "stock_egw",
        "GemiddeldeOppervlakte_2": "avg_area_egw"
    }
)[["RegioS", "year", "stock_egw", "avg_area_egw"]]

# MGW extrahieren
df_mgw = df_wt[df_wt["Woningtype"] == "ZW10340"].rename(
    columns={
        "BeginstandWoningvoorraad_1": "stock_mgw",
        "GemiddeldeOppervlakte_2": "avg_area_mgw"
    }
)[["RegioS", "year", "stock_mgw", "avg_area_mgw"]]

# Merge EGW + MGW
df_structure = df_egw.merge(df_mgw, on=["RegioS", "year"], how="outer")

print(df_structure.shape)
df_structure.head(10)
```

```python
import numpy as np

# Alle relevanten Spalten als Strings behandeln und sauber konvertieren
for col in ["stock_egw", "avg_area_egw", "stock_mgw", "avg_area_mgw"]:
    df_structure[col] = (
        df_structure[col]
        .astype(str)          # alles als String
        .str.strip()          # Leerzeichen entfernen
        .replace(".", np.nan) # "." → NaN
        .replace("", np.nan)  # leere Strings → NaN
        .astype(float)        # in float umwandeln
    )
```

```python
df_structure_step2 = df_structure.copy()

# 1. Bestandsanteile
df_structure_step2["share_egw"] = (
    df_structure_step2["stock_egw"] /
    (df_structure_step2["stock_egw"] + df_structure_step2["stock_mgw"])
)

df_structure_step2["share_mgw"] = 1 - df_structure_step2["share_egw"]

# 2. Verhältnis der Durchschnittsflächen
df_structure_step2["ratio_avg_area"] = (
    df_structure_step2["avg_area_egw"] /
    df_structure_step2["avg_area_mgw"]
)

# 3. Verhältnis der Bestandszahlen
df_structure_step2["ratio_stock"] = (
    df_structure_step2["stock_egw"] /
    df_structure_step2["stock_mgw"]
)

print(df_structure_step2.shape)
df_structure_step2.head(10)
```

```python
df_structure_step3 = df_structure_step2.copy()

df_structure_step3["ratio_price"] = (
    0.6 * df_structure_step3["ratio_avg_area"] +
    0.4 * df_structure_step3["ratio_stock"]
)

print(df_structure_step3.shape)
df_structure_step3.head(10)
```

```python
# Jahr extrahieren in df_verkoop_raw
df_verkoop_raw["year"] = df_verkoop_raw["Perioden"].str[:4].astype(int)

# Merge Struktur + Preise
df_full = df_structure_step3.merge(
    df_verkoop_raw[["RegioS", "year", "GemiddeldeVerkoopprijs_1"]],
    on=["RegioS", "year"],
    how="left"
)

df_full["price_egw"] = (
    df_full["GemiddeldeVerkoopprijs_1"] *
    df_full["ratio_price"] /
    (df_full["share_egw"] * df_full["ratio_price"] + df_full["share_mgw"])
)

df_full["price_mgw"] = df_full["price_egw"] / df_full["ratio_price"]

df_full["ratio_price_realized"] = df_full["price_egw"] / df_full["price_mgw"]

print(df_full.shape)
df_full.head(10)
```

```python
# 1. Kopie von df_full erstellen
df_full_harm = df_full.copy()

# 2. Spalten harmonisieren
df_full_harm = df_full_harm.rename(columns={
    "RegioS": "municipality_code",
    "GemiddeldeVerkoopprijs_1": "price_total",
    "ratio_avg_area": "ratio_area",
    "ratio_price": "ratio_price_model",
    "stock_egw": "housing_stock_egw",
    "stock_mgw": "housing_stock_mgw"
})

# 3. Verhältnis price_egw / price_mgw hinzufügen
df_full_harm["ratio_price_realized"] = df_full_harm["price_egw"] / df_full_harm["price_mgw"]

# 4. Dummy-Gemeinden entfernen
df_full_harm = df_full_harm[df_full_harm["municipality_code"] != "GM0000"]

# 5. Zeilen ohne Strukturinformationen entfernen
df_full_harm = df_full_harm.dropna(subset=["housing_stock_egw", "housing_stock_mgw"])

# 6. Spalten sortieren
df_full_harm = df_full_harm[[
    "municipality_code", "year",
    "price_total", "price_egw", "price_mgw",
    "ratio_price_model", "ratio_price_realized",
    "share_egw", "share_mgw",
    "housing_stock_egw", "housing_stock_mgw",
    "ratio_area", "ratio_stock"
]]

print(df_full_harm.shape)
df_full_harm.head(10)
```
