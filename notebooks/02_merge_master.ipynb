```python
# Klassifikation der NaN‑Fälle (0 vs. echte Lücke)

import pandas as pd

# --- MGW ---
mgw_na = df_master[df_master["new_build_liv_space_mgw"].isna()]

mgw_zero = mgw_na[mgw_na["new_build_stock_mgw"] == 0]
mgw_gap  = mgw_na[mgw_na["new_build_stock_mgw"] > 0]

print("MGW – NaN = 0 (stock = 0):", len(mgw_zero))
print("MGW – NaN = echte Lücke (stock > 0):", len(mgw_gap))

# --- EGW ---
egw_na = df_master[df_master["new_build_liv_space_egw"].isna()]

egw_zero = egw_na[egw_na["new_build_stock_egw"] == 0]
egw_gap  = egw_na[egw_na["new_build_stock_egw"] > 0]

print("EGW – NaN = 0 (stock = 0):", len(egw_zero))
print("EGW – NaN = echte Lücke (stock > 0):", len(egw_gap))
```

```python
# Durchschnittsflächen pro Gemeinde berechnen

# Durchschnittliche Wohnfläche pro Wohnung (EGW und MGW)
df_master["avg_space_egw"] = df_master["living_space_egw"] / df_master["housing_stock_egw_y"]
df_master["avg_space_mgw"] = df_master["living_space_mgw"] / df_master["housing_stock_mgw_y"]
```

```python
# MGW‑NaN mit stock = 0 korrekt auf 0 setzen

# MGW: NaN = 0, wenn stock = 0
mask_mgw_zero = (
    df_master["new_build_liv_space_mgw"].isna() &
    (df_master["new_build_stock_mgw"] == 0)
)

df_master.loc[mask_mgw_zero, "new_build_liv_space_mgw"] = 0
```

```python
# 135 Lücken korrekt imputieren (MGW: 121 echte Lücken / EGW: 14 echte Lücken)

# MGW: echte Lücken (stock > 0)
mask_mgw_gap = (
    df_master["new_build_liv_space_mgw"].isna() &
    (df_master["new_build_stock_mgw"] > 0)
)

df_master.loc[mask_mgw_gap, "new_build_liv_space_mgw"] = (
    df_master.loc[mask_mgw_gap, "new_build_stock_mgw"] *
    df_master.loc[mask_mgw_gap, "avg_space_mgw"]
)

# EGW: echte Lücken (stock > 0)
mask_egw_gap = (
    df_master["new_build_liv_space_egw"].isna() &
    (df_master["new_build_stock_egw"] > 0)
)

df_master.loc[mask_egw_gap, "new_build_liv_space_egw"] = (
    df_master.loc[mask_egw_gap, "new_build_stock_egw"] *
    df_master.loc[mask_egw_gap, "avg_space_egw"]
)
```

```python
df_master.isna().sum().sort_values(ascending=False).head(10)
```

```python
# 3 Total‑NaN korrekt berechnen

# Fehlende Total-Werte berechnen (MGW + EGW)
mask_tot_na = df_master["new_build_liv_space_tot"].isna()

df_master.loc[mask_tot_na, "new_build_liv_space_tot"] = (
    df_master.loc[mask_tot_na, "new_build_liv_space_mgw"] +
    df_master.loc[mask_tot_na, "new_build_liv_space_egw"]
)
```

```python
# Preis‑NaN interpolieren

# Preisvariablen, die interpoliert werden sollen
price_cols = [
    "price_mgw",
    "price_egw",
    "ratio_price_model",
    "ratio_price_realized",
    "ratio_area"
]

# Sortieren für zeitliche Reihenfolge
df_master = df_master.sort_values(["municipality_code", "year"])

# Interpolation pro Gemeinde, auch am Rand
for col in price_cols:
    df_master[col] = (
        df_master
        .groupby("municipality_code")[col]
        .apply(lambda g: g.interpolate(method="linear", limit_direction="both", limit_area="outside"))
        .reset_index(level=0, drop=True)
    )
```
