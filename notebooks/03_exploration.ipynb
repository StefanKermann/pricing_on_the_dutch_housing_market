```python
############################################################################################
# 3.2. Preisverteilungen visualisieren: price_total, price_egw, price_mgw auf Gemeindeebeene
############################################################################################

import matplotlib.ticker as mtick
import numpy as np

# Histogramm der Preisvariablen mit festen Achsen

fig, axes = plt.subplots(1, 3, figsize=(22, 6))

# Einheitliche Achsenlimits
x_min, x_max = 0, 1_400_000
y_min, y_max = 0, 200

# Einheitliches Raster für X-Achse (alle 100.000 EUR)
xticks = np.arange(x_min, x_max + 1, 100_000)

# Plot 1: price_total
sns.histplot(df_master["price_total"], kde=True, ax=axes[0], color="steelblue")
axes[0].set_title("Verteilung: price_total")
axes[0].set_xlabel("Preis (EUR)")
axes[0].set_xlim(x_min, x_max)
axes[0].set_ylim(y_min, y_max)
axes[0].set_xticks(xticks)
axes[0].xaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

# Plot 2: price_egw
sns.histplot(df_master["price_egw"], kde=True, ax=axes[1], color="darkgreen")
axes[1].set_title("Verteilung: price_egw")
axes[1].set_xlabel("Preis (EUR)")
axes[1].set_xlim(x_min, x_max)
axes[1].set_ylim(y_min, y_max)
axes[1].set_xticks(xticks)
axes[1].xaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

# Plot 3: price_mgw
sns.histplot(df_master["price_mgw"], kde=True, ax=axes[2], color="darkred")
axes[2].set_title("Verteilung: price_mgw")
axes[2].set_xlabel("Preis (EUR)")
axes[2].set_xlim(x_min, x_max)
axes[2].set_ylim(y_min, y_max)
axes[2].set_xticks(xticks)
axes[2].xaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

plt.tight_layout()
plt.show()
```

```python
import matplotlib.ticker as mtick

# Boxplots der Preisvariablen mit festen Achsen

fig, axes = plt.subplots(1, 3, figsize=(22, 6))

# Einheitliche Y-Achse
y_min, y_max = 0, 1_400_000

# Plot 1: price_total
sns.boxplot(y=df_master["price_total"], ax=axes[0], color="steelblue")
axes[0].set_title("Boxplot: price_total")
axes[0].set_ylim(y_min, y_max)
axes[0].yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

# Plot 2: price_egw
sns.boxplot(y=df_master["price_egw"], ax=axes[1], color="darkgreen")
axes[1].set_title("Boxplot: price_egw")
axes[1].set_ylim(y_min, y_max)
axes[1].yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

# Plot 3: price_mgw
sns.boxplot(y=df_master["price_mgw"], ax=axes[2], color="darkred")
axes[2].set_title("Boxplot: price_mgw")
axes[2].set_ylim(y_min, y_max)
axes[2].yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))

plt.tight_layout()
plt.show()
```

```python
################################################################################
# 3.4.1 Preisverteilungen visualisieren: price_total auf Provinzebeene (barplot)
################################################################################

import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

avg_prices = (
    df_master.groupby("province")["price_total"]
    .mean()
    .sort_values()
)

plt.figure(figsize=(14, 7))
avg_prices.plot(kind="barh", color="steelblue")

plt.title("Average purchase price per province (2019–2023)")
plt.xlabel("Price (EUR)")
plt.ylabel("Province")
plt.gca().xaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))
plt.tight_layout()
plt.show()
```

```python
##################################################################
# 3.6. Verhältnis EGW/MGW pro Provinz (Preisrelationen) (barplot)
##################################################################

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick

# Korrektes Verhältnis: EGW / MGW
df_master["ratio_egw_mgw"] = df_master["price_egw"] / df_master["price_mgw"]

ratio = (
    df_master.groupby("province")["ratio_egw_mgw"]
    .mean()
    .sort_values()
)

plt.figure(figsize=(14, 6))

sns.barplot(
    x=ratio.index,
    y=ratio.values,
    palette="viridis"
)

plt.title("Price-relation EGW/MGW per Province (Avg. 2019–2023)")
plt.xlabel("Province")
plt.ylabel("Price-ratio (EGW / MGW)")

plt.xticks(rotation=45)
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:.2f}'))

plt.tight_layout()
plt.show()
```

```python
#################################################
# 3.8.3 Preisänderung 2019-2023 auf Gemeindeebene
#################################################

import matplotlib.pyplot as plt
import numpy as np

# Preisänderung berechnen
change = (
    df_master.groupby(['gemeentenaam', 'year'])['price_total']
    .median()
    .unstack()
)

change['diff_19_23'] = change[2023] - change[2019]
change_sorted = change['diff_19_23'].sort_values()

# Top 20 & Bottom 20
top20 = change_sorted.tail(20)
bottom20 = change_sorted.head(20)

xticks = np.arange(0, 300001, 50000)

# Top 20
plt.figure(figsize=(10,6))
plt.barh(top20.index, top20.values, color='red')
plt.title('Municipality – Price Change 2019 → 2023 (Top 20)', fontsize=12)
plt.xlabel('Change in EUR', fontsize=10)
plt.ylabel('Municipality', fontsize=10)
plt.xlim(0, 300000)
plt.xticks(xticks, fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()

# Bottom 20
plt.figure(figsize=(10,6))
plt.barh(bottom20.index, bottom20.values, color='green')
plt.title('Municipality – Price Change 2019 → 2023 (Bottom 20)', fontsize=12)
plt.xlabel('Change in EUR', fontsize=10)
plt.ylabel('Municipality', fontsize=10)
plt.xlim(0, 300000)
plt.xticks(xticks, fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
#####################
# 3.10 Clusteranalyse
#####################

# Variablenauswahl für die Clusteranalyse

cluster_vars = [
    'price_total',
    'income_mean',
    'woz_mean',
    'bevolkingsdichtheid',
    'avg_space_egw',
    'ratio_egw_mgw'
]

df_cluster = df_master.groupby('gemeentenaam')[cluster_vars].median()
```

```python
# Standardisierung (Grundlage für K‑Means)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cluster)
```

```python
# Bestimmung der Anzahl Cluster

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []

for k in range(2, 11):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_scaled)
    inertia.append(km.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(2, 11), inertia, marker='o')
plt.title('Elbow Method – Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# K‑Means‑Clustering durchführen

k = 4  # falls du eine andere Zahl willst, einfach ändern

kmeans = KMeans(n_clusters=k, random_state=42)
df_cluster['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# Cluster visualisieren

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
components = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
plt.scatter(components[:,0], components[:,1], c=df_cluster['cluster'], cmap='tab10', s=20)
plt.title('Municipality Clusters (PCA 2D)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd
import pyogrio

file_path = '../data_raw/wijkenbuurten_2023_v3.gpkg'

# 1. Layer automatisch finden (sucht nach 'gemeente')
layers = pyogrio.list_layers(file_path)
target_layer = [l for l in layers[:, 0] if 'gemeente' in l.lower()][0]

# 2. Layer laden und in WGS84 transformieren (für Lat/Lon)
gdf = gpd.read_file(file_path, layer=target_layer)
gdf_wgs84 = gdf.to_crs(epsg=4326)

# 3. Zentroide (Mittelpunkte) berechnen
gdf_wgs84['longitude'] = gdf_wgs84.geometry.centroid.x
gdf_wgs84['latitude'] = gdf_wgs84.geometry.centroid.y

# 4. Flexibles Mapping für CBS-Spaltennamen (Code und Name)
# CBS nutzt je nach Export 'statcode', 'RegioS', 'gemeentecode' oder 'gm_code'
potential_code_cols = ['statcode', 'RegioS', 'gemeentecode', 'gm_code']
potential_name_cols = ['statnaam', 'gemeentenaam', 'gm_naam']

# Finde die erste existierende Spalte aus den Listen
code_col = next((c for c in potential_code_cols if c in gdf_wgs84.columns), None)
name_col = next((c for c in potential_name_cols if c in gdf_wgs84.columns), None)

if not code_col:
    raise KeyError(f"Keine Code-Spalte gefunden. Vorhanden: {gdf_wgs84.columns.tolist()}")

# 5. Daten bereinigen: 'GM' entfernen (aus 'GM0363' wird '0363')
gdf_wgs84['gemeindecode_cbs'] = gdf_wgs84[code_col].str.replace('GM', '', regex=False)

# 6. Finale Tabelle erstellen
result = gdf_wgs84[['gemeindecode_cbs', name_col, 'longitude', 'latitude']]
result = result.rename(columns={name_col: 'gemeindename'})

# 7. Export als CSV
result.to_csv('cbs_gemeinden_niederlande_2023.csv', index=False)

print(f"Erfolg! Layer '{target_layer}' verarbeitet.")
print(f"Spalte '{code_col}' wurde als CBS-Code verwendet.")
print(result.head())
```

```python
###############################################################
# 3.12. Ermittlung und Visualisierung statistischer Extremwerte 
###############################################################

from scipy import stats
import numpy as np

df_geo['z_price_total'] = stats.zscore(df_geo['price_total'])

outliers_z = df_geo[np.abs(df_geo['z_price_total']) > 2.5]

outliers_z[['gemeentenaam', 'price_total', 'z_price_total']]
```

```python
Q1 = df_geo['price_total'].quantile(0.25)
Q3 = df_geo['price_total'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers_iqr = df_geo[(df_geo['price_total'] < lower) | (df_geo['price_total'] > upper)]

outliers_iqr[['gemeentenaam', 'price_total']]
```

```python
# Grafische Darstellung Extremwerte 'price_total' auf Basis von Geo-Koordinaten

plt.figure(figsize=(8,10))

# Alle Gemeinden
plt.scatter(
    df_geo['longitude'],
    df_geo['latitude'],
    c='lightgrey',
    s=20,
    alpha=0.5,
    label='All Municipalities'
)

# Outlier (Z-Score)
plt.scatter(
    outliers_z['longitude'],
    outliers_z['latitude'],
    c='red',
    s=40,
    alpha=0.5,
    label='Outliers (Z-Score)'
)

plt.title('Outlier Municipalities (Price Total)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Vorbereitung grafische Darstellung weiterer Extremwerte auf Basis von Geo-Koordinaten

from scipy import stats

variables = ['income_mean', 'woz_mean', 'bevolkingsdichtheid']
threshold = 2.5

for var in variables:
    df_geo[f'z_{var}'] = stats.zscore(df_geo[var])
    
```

```python
import matplotlib.pyplot as plt

def plot_outlier_map(df, var, threshold=2.5):
    outliers = df[abs(df[f'z_{var}']) > threshold]

    plt.figure(figsize=(8,10))

    # Alle Gemeinden (grau)
    plt.scatter(
        df['longitude'],
        df['latitude'],
        c='lightgrey',
        s=20,
        alpha=0.5,
        label='All Municipalities'
    )

    # Outlier (rot)
    plt.scatter(
        outliers['longitude'],
        outliers['latitude'],
        c='red',
        s=40,
        alpha=0.5,
        label=f'Outliers ({var})'
    )

    plt.title(f'Outlier Map: {var} (|z| > {threshold})')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# Grafische Darstellung Extremwerte 'income_mean' auf Basis von Geo-Koordinaten

plot_outlier_map(df_geo, 'income_mean', threshold=2.5)
```

```python
# Grafische Darstellung Extremwerte 'woz_mean' auf Basis von Geo-Koordinaten

plot_outlier_map(df_geo, 'woz_mean', threshold=2.5)
```

```python
# Grafische Darstellung Extremwerte 'bevolkingsdichtheid' auf Basis von Geo-Koordinaten

plot_outlier_map(df_geo, 'bevolkingsdichtheid', threshold=2.5)
```
